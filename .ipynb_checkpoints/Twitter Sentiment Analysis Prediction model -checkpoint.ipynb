```python
dataset['tweet'] = dataset['tweet'].str.replace('!"#$%&\'+,-./:@\]^_`{|}~','')
print(dataset['tweet'].head())
```

```python
import nltk 
nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords = stopwords.words('english')
from nltk.tokenize import word_tokenize
nltk.download('punkt')
```

```python
dataset['tweet_without_stopwords'] = dataset['tweet'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stopwords)]))
```

```python
dataset.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
tweets_countvectorizer = vectorizer.fit_transform(dataset['tweet'])
```

```python
print(vectorizer.get_feature_names())
```

```python
print(tweets_countvectorizer.toarray())  
```

```python
tweets = pd.DataFrame(tweets_countvectorizer.toarray())
```

```python
tweets
```

```python
x = tweets
y = dataset["label"]
print(x.shape , y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
```

```python
from sklearn.naive_bayes import MultinomialNB
NB_classifier = MultinomialNB()
NB_classifier.fit(X_train, y_train)
```

```python
# Predicting the Test set results
y_pred = NB_classifier.predict(X_test)
```
